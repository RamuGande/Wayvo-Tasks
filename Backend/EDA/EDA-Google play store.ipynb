```python
#Here some sizes are in mb and some are in kb so convert all the data into kb.
#1MB = 1024KB
def sizeprocess(item):
    if str(item)[-1] == 'M':
        res = float(str(item).replace('M',''))
        res = res*1024
        return res
    elif str(item)[-1]=='k':
        res = str(item).replace('k','')
        return res
    else :
        return str(np.nan)
```

```python
df_copy['Size'] = df_copy['Size'].apply(sizeprocess)
```

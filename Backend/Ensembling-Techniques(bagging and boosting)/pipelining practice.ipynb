```python
from sklearn.impute import SimpleImputer
k=SimpleImputer(strategy='most_frequent')
df[cat_cols] = k.fit_transform(df[cat_cols])#replaces all the null values fot cat_cols with theirrespective mode values
```

```python
def outlier(col):
    print(col)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1
    lower_fence = Q1-1.5*IQR
    upper_fence = Q3+1.5*IQR
    print(lower_fence)
    print(upper_fence)
```

```python
for col in num_cols:
    outlier(col)
```

```python
sns.boxplot(df[num_cols])
#plt.boxplot(df[num_cols],labels = num_cols)
#plt.ylim(-10,150)
plt.figure(figsize=(12,100))
plt.show()
```

```python
i=1
plt.figure(figsize=(12,12))
for col in num_cols:
    Q1 = np.quantile(df[col],0.25)
    Q3 = np.quantile(df[col],0.75)
    IQR = Q3-Q1
    lower_fence = Q1-1.5*IQR
    upper_fence = Q3+1.5*IQR
    df[col] = np.where((df[col]<lower_fence) | (df[col]>upper_fence),df[col].median(),df[col])
    plt.xlabel(col)
    plt.subplot(2,3,i)
    i=i+1
    sns.boxplot(df[col])
```

```python
df['fare'] = np.where(df.fare<lower_fence,lower_fence,np.where(df.fare>upper_fence,upper_fence,df['fare']))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_samples = 1000,n_features = 2 ,noise = 3.0,n_targets = 1)
```

```python
x
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue = y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel = "linear")
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred=svr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import  GridSearchCV
params = {'C' :[0.1,0.2,1,2,3,50,100,10],
          'gamma':[1,0.1,0.2,0.001,0.003],
          'kernel':['linear'],
          'epsilon':[0.01,0.1,0.2,0.3]
         }
grid = GridSearchCV(svr,param_grid = params,cv=5,verbose = 3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

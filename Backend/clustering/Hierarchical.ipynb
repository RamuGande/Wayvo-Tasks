```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn import datasets
data = datasets.load_iris()
```

```python
x = data.data
y = data.target
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
#applying pca to reduce the dimensionality
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca_scaled = pca.fit_transform(x_scaled)
```

```python
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=y)
plt.show()
```

```python
#agglomerative clustering
```

```python
import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20,7))
plt.title("dendrogram")
sc.dendrogram(sc.linkage(pca_scaled,method = "ward"))
plt.xlabel("sample data")
plt.ylabel("euclidean_distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters = 2,linkage = "ward",affinity="euclidean")
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
from sklearn.metrics import silhouette_score
```

```python
scores=[]
for i in range(2,11):
    agglo = AgglomerativeClustering(n_clusters = i,linkage = "ward",affinity="euclidean")
    agglo.fit(pca_scaled)
    score = silhouette_score(pca_scaled,agglo.labels_)
    scores.append(score)
```

```python
scores
```

```python
plt.plot(range(2,11),scores)
plt.xlabel("no of clusters")
plt.xticks(range(2,11))
plt.ylabel("silhouette score")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
x,y = make_blobs(n_samples=1000,n_features = 2,centers = 3)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size = 0.2)
```

```python
plt.scatter(x[:,0],x[:,1],c = y)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3,init = 'k-means++')
```

```python
y_pred = model.fit_predict(x_train)
```

```python
y_pred
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_pred)
```

```python
#we can change no.of clusters and check
```

```python
#silhouette score is an evaluation metric for clustering algorithms
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x_train,model.labels_)#y_pred = model.labels
```

```python
silhouette_score(x_train,y_pred)
```

```python
#how to select optimal k (centriod) >> elbow method.
```

```python
model.inertia_#this is wcss score when k is 3
```

```python
wcss=[]
for i in range(1,10):
    model = KMeans(n_clusters=i)
    model.fit(x_train)
    wcss.append(model.inertia_)
```

```python
wcss
```

```python
#plot the elbow curve
plt.plot(range(1,10),wcss)
plt.xlabel("No.of clusters")
plt.ylabel("Wcss")
plt.xticks(range(1,10))
plt.show()
```

```python
#based on thr plot the optimal k/centriod is 3
```

```python
silhouette_scores = []
for i in range(2,10):
    model = KMeans(n_clusters = i)
    model.fit(x_train)
    score = silhouette_score(x_train,model.labels_)
    silhouette_scores.append(score)
silhouette_scores
```

```python
plt.plot(range(2,10),silhouette_scores)
plt.xlabel("no.of clusters")
plt.xticks(range(2,10))
plt.ylabel("silhouette score")
plt.show()
```

```python
#everytime its better to compare with both wcss scores and silhouette scores for optimal k
```
